```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Part 1:
roll_number = 102317080
np.random.seed(roll_number)
sales_data = np.random.randint(1000, 5000, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)

# Part 2:

#  Display first 5 rows
print(df.head())
print("\nSummary Statistics:\n", df.describe())
df['Total Sales'] = df.sum(axis=1)
total_sales_per_category = df[categories].sum()
print("\nTotal Sales per Category:\n", total_sales_per_category)
print("\nTotal Sales per Month:\n", df['Total Sales'])

#  Calculate average sales growth between consecutive months
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df['Growth Rate'].fillna(0, inplace=True)

# Apply Discount  Roll Number ke basis par
if roll_number % 2 == 0:
    df['Electronics'] *= 0.90  # 10% Discount
else:
    df['Clothing'] *= 0.85  # 15% Discount

print("\nDataFrame after Discount:\n", df)

# Part3:

# Line Plot for Monthly Sales Trends
df[categories].plot(kind='line', figsize=(10, 5), marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

#Box Plot for Sales Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

# (i)
abs_values = np.abs(array)
print("Absolute Values:\n", abs_values)

# (ii)
percentiles = {
    "25th": np.percentile(array, 25, axis=None),
    "50th (Median)": np.percentile(array, 50, axis=None),
    "75th": np.percentile(array, 75, axis=None)
}
print("\nPercentiles:", percentiles)

# (iii)
print("\nMean:", np.mean(array))
print("Median:", np.median(array))
print("Standard Deviation:", np.std(array))

# Column-wise
print("\nColumn-wise Mean:", np.mean(array, axis=0))
print("Column-wise Median:", np.median(array, axis=0))
print("Column-wise Standard Deviation:", np.std(array, axis=0))

# Row-wise
print("\nRow-wise Mean:", np.mean(array, axis=1))
print("Row-wise Median:", np.median(array, axis=1))
print("Row-wise Standard Deviation:", np.std(array, axis=1))
```

```python
def swap_list_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

# Example
lst = [10, 20, 30, 40, 50]
index1, index2 = 1, 3  # Swap elements at index 1 and 3
print("Before Swap:", lst)
print("After Swap:", swap_list_elements(lst, index1, index2))
```

```python
def swap_set_elements(s, elem1, elem2):
    lst = list(s)
    index1, index2 = lst.index(elem1), lst.index(elem2)

    # Swap elements
    lst[index1], lst[index2] = lst[index2], lst[index1]

    return set(lst)

# Example
s = {1, 2, 3, 4, 5}
elem1, elem2 = 2, 4  # Swap these elements
print("Before Swap:", s)
print("After Swap:", swap_set_elements(s, elem1, elem2))

```
